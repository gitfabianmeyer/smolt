```python
# probas
# terminologie xxy : from node x to node x using transistion y
epsV = .2
epsN = .8
nna = .2
nnb = .5
nva = .1
nvb = .2
vna = .2
vnb = .1
vva = .5
vvb = .2

# using b
v2 = epsV*vvb + epsN *nvb
n2 = epsV*vnb + epsN *nnb
# using b
v3 = v2* vvb + n2 *nvb
n3 = v2* vnb + n2*nnb
# using b
v4 = v3 * vvb + n3 *nvb
n4 = v3 * vnb + n3 *nnb
# using a
v5 = v4 * vva + n4 *nva
n5 = v4 * vna + n4 * nna

resForward = v5 + n5
```

```python
a = Digraph()
a.attr(rankdir = 'LR')

a.node('v1', 'V: ' + str(epsV), shape='doublecircle')
a.node('v2', 'V: ' + str(v2))
a.node('v3', 'V: ' + str(v3))
a.node('v4', 'V: ' + str(v4))
a.node('v5', 'V: ' + str(v5))

a.node('n1', 'N: '+ str(epsN),  shape='doublecircle')
a.node('n2', 'N: ' + str(n2))
a.node('n3', 'N: ' + str(n3))
a.node('n4', 'N: ' + str(n4))
a.node('n5', 'N: ' + str(n5))

# first letter: b
a.edge('v1', 'v2', label = str(vvb))
a.edge('v1', 'n2', label = str(vnb))
a.edge('n1', 'n2', label = str(nnb))
a.edge('n1', 'v2', label= str(nvb))

# b
a.edge('v2', 'v3', label = str(vvb))
a.edge('v2', 'n3', label = str(vnb))
a.edge('n2', 'n3', label = str(nnb))
a.edge('n2', 'v3', label= str(nvb))

# b
a.edge('v3', 'v4', label = str(vvb))
a.edge('v3', 'n4', label = str(vnb))
a.edge('n3', 'n4', label = str(nnb))
a.edge('n3', 'v4', label= str(nvb))

# a
a.edge('v4', 'v5', label = str(vva))
a.edge('v4', 'n5', label = str(vna))
a.edge('n4', 'n5', label = str(nna))
a.edge('n4', 'v5', label = str(nva))
a
```
