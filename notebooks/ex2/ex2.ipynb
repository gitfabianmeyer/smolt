2.1 FST

Keep the first letter of the name, and drop all occurrences of non-initial a, e, h, i, o, u, w, y. In FSTs, you can abbreviate this class with V.  

Comment on the solution: I add a # as end-of-word symbol as well as an &alpha; as intermediate marker for deleters, which will be used for part c)

```python
a = Digraph()
a.graph_attr['rankdir'] = 'LR'
a.node('0')
a.node('1')
a.node('2', shape='doublecircle')
# read first letter and output same letter
a.edge('0', '1', label='&Sigma; : &Sigma;')

# read letter from V and output interm. state &alpha; to remove it later
a.edge('1', '1', label='V : &alpha;')
# read any letter from Alphabet not in V and print same letter. We we just add one graph for convenience
a.edge('1', '1', label=r'(&Epsilon; \\ V) : (&Epsilon; \\ V)')

# read end-of-word token and print nothing
a.edge('1', '2', label='# : &epsilon;')
print('Keep first letter, remove (her: replace with &alpha; ) every letter in V')
a
```

```python
print('b) Replace every letter with respective number, keep first')
b = Digraph()
b.graph_attr['rankdir'] = 'LR'
b.node('3')
b.node('4')
b.node('5', shape='doublecircle')

#read first and output same
b.edge('3', '4', label='&Sigma; : &Sigma;')

# read letter and print out respective Group Cx. We use Cx as abbrevation to not put an edge for every letter here
b.edge('4', '4', label='C1 : 1')
b.edge('4', '4', label = 'C2 : 2')
b.edge('4', '4', label = 'C3 : 3')
b.edge('4', '4', label = 'C4 : 4')
b.edge('4', '4', label = 'C5 : 5')
b.edge('4', '4', label = 'C6 : 6')
b.edge('4', '5', label = '# : &epsilon;')
b
```

```python
print('c) replace sequences of identical numbers if they were adjacent in original and remove intermidiate symbol from a)')

c = Digraph()
c.graph_attr["rankdir"] = 'LR'
c.node('6')
c.node('7')
c.node('14', shape='doublecircle')

# read and output first as always
c.edge('6', '7', label='&Sigma; : &Sigma;')

# read number from b), print letter from b), if number comes again: print nothing and return to same state to delete repetitions
c.node('8')
c.edge('7', '8', label='1 : 1')
c.edge('8', '8', label = '1 : &epsilon;')
c.node('9')
c.edge('7', '9', label='2 : 2')
c.edge('9','9', label= '2 : &epsilon;')
c.node('10')
c.edge('7', '10', label='3 : 3')
c.edge('10', '10', label= '3 : &epsilon;')
c.node('11')
c.edge('7', '11', label='4 : 4')
c.edge('11', '11', label= '4: &epsilon;')
c.node('12')
c.edge('7', '12', label='5 : 5')
c.edge('12', '12', label='5 : &epsilon;')
c.node('13')
c.edge('7', '13', label='6 : 6')
c.edge('13', '13', label='6: &epsilon;')


# read interm outout nothing to delete it
c.edge('7', '7', label='&alpha; : &epsilon;')
c.edge('8', '7', label='&alpha; : &epsilon;')
c.edge('9', '7', label='&alpha; : &epsilon;')
c.edge('10', '7', label='&alpha; : &epsilon;')
c.edge('11', '7', label='&alpha; : &epsilon;')
c.edge('12', '7', label='&alpha; : &epsilon;')
c.edge('13', '7', label='&alpha; : &epsilon;')

# read any other symbol than the same as before, to respective state
# 1x1 to ...
c.edge('8', '9', label ='2 : 2', color='red')
c.edge('8', '10', label ='3 : 3', color='red')
c.edge('8', '11', label ='4 : 4', color='red')
c.edge('8', '12', label ='5 : 5', color='red')
c.edge('8', '13', label ='6 : 6', color='red')

# 1x2 to ...
c.edge('9', '8', label ='1 : 1', color='blue')
c.edge('9', '10', label ='3 : 3', color='blue')
c.edge('9', '11', label ='4 : 4', color='blue')
c.edge('9', '12', label ='5 : 5', color='blue')
c.edge('9', '13', label ='6 : 6', color='blue')

# 1x3 to ...
c.edge('10', '8', label ='1 : 1', color='green')
c.edge('10', '9', label ='2 : 2', color='green')
c.edge('10', '11', label ='4 : 4', color='green')
c.edge('10', '12', label ='5 : 5', color='green')
c.edge('10', '13', label ='6 : 6', color='green')

# 1x4 to ...
c.edge('11', '8', label ='1 : 1', color='yellow')
c.edge('11', '9', label ='2 : 2', color='yellow')
c.edge('11', '10', label ='3 : 3', color='yellow')
c.edge('11', '12', label ='5 : 5', color='yellow')
c.edge('11', '13', label ='6 : 6', color='yellow')

# 1x5 to ...
c.edge('12', '8', label ='1 : 1', color='brown')
c.edge('12', '9', label ='2 : 2', color='brown')
c.edge('12', '10', label ='3 : 3', color='brown')
c.edge('12', '11', label ='4 : 4', color='brown')
c.edge('12', '13', label ='6 : 6', color='brown')

# 1x5 to ...
c.edge('13', '8', label ='1 : 1', color='purple')
c.edge('13', '9', label ='2 : 2', color='purple')
c.edge('13', '10', label ='3 : 3', color='purple')
c.edge('13', '11', label ='4 : 4', color='purple')
c.edge('13', '12', label ='5 : 5', color='purple') 

# and from every state to final
c.edge('7', '14', label='# : &epsilon;')
c.edge('8', '14', label='# : &epsilon;')
c.edge('9', '14', label='# : &epsilon;')
c.edge('10', '14', label='# : &epsilon;')
c.edge('11', '14', label='# : &epsilon;')
c.edge('12', '14', label='# : &epsilon;')
c.edge('13', '14', label='# : &epsilon;')
c
```

```python
print('d) convert to LDDD by dropping or padding with zeros')

d = Digraph()
d.graph_attr['rankdir'] = 'LR'

d.node('15')
d.node('16')
d.node('17')
d.node('18')
d.node('19')
d.node('20')
d.node('21',  shape='doublecircle')
d.node('22', shape='doublecircle')

# read D out D
d.edge('15', '16', label='D : D')

# the already perfect path
d.edge('16', '17', label='L : L')
d.edge('17', '20', label='L : L')
d.edge('20', '21', label='L : L')


# only on D, then # has to be padded
d.edge('16', '18', label='# : 0')
d.edge('18', '19', label='&epsilon; : 0')
d.edge('19', '22', label='&epsilon; : 0')

# eos after first L, second L...
d.edge('17', '19', label = '# : 0')
d.edge('20', '22', label = '# : 0')

# anythin after DLLL: read and output eps
d.edge('21', '21', label='L : &epsilon;')
d.edge('21', '21', label='# : &epsilon;')
d
```
